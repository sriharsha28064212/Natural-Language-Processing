```python
def gender_features(word):
    return{'last_letter': word[-1]}
```

```python
gender_features('harsha')
```

```python
from nltk.corpus import names
```

```python
labeled_names = [(name, 'male') for name in names.words('male.txt')] + [(name,'female') for name in names.words('female.txt')]
```

```python
import random
random.shuffle(labeled_names)
feature_sets = [(gender_features(n), gender)
                    for (n, gender) in labeled_names]
train_set, test_set = feature_sets[500:], feature_sets[:500]
```

```python
feature_sets
```

```python
import nltk
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
g1 = classifier.classify(gender_features('Rohit'))
g2 = classifier.classify(gender_features('Shreya'))
print("{} is most probably a {}.".format('Rohit',g1))
print("{} is most probably a {}.".format('Shreya',g2))
```

```python
# Test the accuracy of the classifier on the test data
print("\n Accuracy of the model is :",nltk.classify.accuracy(classifier, test_set)*100,"%\n")
```
